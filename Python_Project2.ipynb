# QUESTION NO 07:
 Being an Analyst understands the total number of purchases made through different channels
which can help find the percentage these channels contribute.

a. Find out the total purchases done by a customer through different channels.

b. Display the percentage of the store and other channels’ contribution to the total
purchases.

```python
df.columns
```

```python
df['Total_Purchase']=df['NumWebPurchases']+df['NumCatalogPurchases']+df['NumStorePurchases']
df.head()
```

```python
# Percentage of store purchase
df["store_pur"]=(df["NumWebPurchases"]/df["Total_Purchase"])*100
df["store_pur"]
```

```python
# Percentage of other channel purchase
df['pur_other']=((df["NumCatalogPurchases"]+df["NumWebPurchases"])/df["Total_Purchase"])*100
df['pur_other']
```

```python
# Display the percentage
df.filter(regex='Store|Total_P|store_pur|pur_other',axis=1)
```

# Question no 08:
The marketing manager wants to understand the performance of different marketing
